```python
import tensorflow as tf
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dropout
from keras.layers import Flatten
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_train.reshape(-1, 28, 28, 1)
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes = 10)
y_test = keras.utils.to_categorical(y_train, num_classes = 10)
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(60):
    plt.imshow(x_train[i].reshape( 28,28))
    plt.show()
```

```python
model=Sequential()
model.add(Conv2D(8, (3, 3), activation = 'relu', input_shape = (28, 28, 1)))
model.add(Flatten())
model.add(Dense(10, activation = 'softmax'))
model.summary()
```

```python
model.compile(loss = keras.losses.categorical_crossentropy, optimizer = keras.optimizers.Adadelta())
```

```python
model.fit(x_train, y_train, epochs = 5, batch_size = 64)
```

```python
img = x_train[2].reshape((28, 28,))
plt.imshow(img)
```

```python
import numpy as np
out = model.predict(img.reshape(-1, 28, 28, 1))
print("The model thinks this is", np.argmax(out))
```
